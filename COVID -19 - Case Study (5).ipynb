# Exploring Egypt Data

```python
Covid_World_Over_Time = Covid_World.groupby('date') [['total_cases','new_cases','total_deaths','new_deaths']].sum().reset_index()
Covid_World_Over_Time
```

```python
Covid_World_Over_Time['date'] = pd.to_datetime(Covid_World_Over_Time['date'])
```

```python
Covid_World_Over_Time.info()
```

```python
plt.figure(figsize=(20,10))
plt.plot(Covid_World_Over_Time['date'],Covid_World_Over_Time['total_cases'])
plt.xlabel('Months')
plt.ylabel('Total Cases')
plt.title('Total Cases All Over The world')
```

```python
plt.figure(figsize=(20,10))
plt.plot(Covid_World_Over_Time['date'],Covid_World_Over_Time['total_deaths'])
plt.xlabel('Months')
plt.ylabel('Total Deaths')
plt.title('Total Cases All Over The world')
```

```python
plt.figure(figsize=(20,10))
plt.plot(Covid_World_Over_Time['date'],Covid_World_Over_Time['new_cases'], marker='o', c='r', lw='2')
plt.xlabel('Months')
plt.ylabel('New Cases')
plt.title('Total Cases All Over The world')
```

```python
plt.figure(figsize=(20,10))
# plt.plot(Covid_World_Over_Time['date'],Covid_World_Over_Time['total_cases'])
plt.plot(Covid_World_Over_Time['date'],Covid_World_Over_Time['total_deaths'], label='Deaths')
plt.plot(Covid_World_Over_Time['date'],Covid_World_Over_Time['new_cases'], label='New')
plt.legend(loc=0)
```
